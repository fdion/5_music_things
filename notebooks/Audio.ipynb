# IPython.display.Audio

So... is this a visual or auditive thing?

http://ipython.org/ipython-doc/stable/api/generated/IPython.display.html#IPython.display.Audio

```python
import numpy as np
from IPython.display import Audio
```

```python
sample_rate = 44100  # Hz or cps, F = 1/T
t = np.linspace(0, 5, sample_rate*5)  # 5 seconds

# from the example in the above link:
left = np.sin(2*np.pi*220*t)
right = np.sin(2*np.pi*224*t)
Audio([left, right],rate=sample_rate)
```
